# Mean and Standard Deviation 

## Global

```python
import numpy as np
import pandas as pd
import math

anom=pd.read_csv('temperature-anomaly.csv')
anom.drop(['Code'],axis=1,inplace=True)

globe=anom.loc[lambda anom: anom['Entity'] == 'Global']

#1850-1900
a=0
for i in range(0,50):
    a=a+globe.iloc[i,2]
#mean&std1
m1=a/len(range(0,50))

sum1=0
for i in range(0,50):
    sum1=sum1+((globe.iloc[i,2]-m1)**2)

std1=math.sqrt(sum1/(len(range(0,50))-1))
    
#1901-1950
b=0
for i in range(51,100):
    b=b+globe.iloc[i,2]

#mean&std2
m2=b/len(range(51,100))

sum2=0
for i in range(51,100):
    sum2=sum2+((globe.iloc[i,2]-m2)**2)

std2=math.sqrt(sum2/(len(range(51,100))-1))

#1951-2018('present')
c=0
for i in range(101,169):
    c=c+globe.iloc[i,2]

#mean&std3
m3=c/len(range(101,169))

sum3=0
for i in range(101,169):
    sum3=sum3+((globe.iloc[i,2]-m3)**2)

std3=math.sqrt(sum3/(len(range(101,169))-1))


print("the mean of the global median temperature anomalies (C) from 1850 to 1900 is %f with a standard deviation of %f" %(m1,std1))
print("the mean of the global median temperature anomalies (C) from 1901 to 1950 is %f with a standard deviation of %f" %(m2,std2))
print("the mean of the global median temperature anomalies (C) from 1951 to 2018 (present) is %f with a standard deviation of %f\n" %(m3,std3))
```

```python
import matplotlib.pyplot as plt
from matplotlib import pylab
import numpy as np
from scipy.optimize import curve_fit

fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,10))
ax.set_title('Southern Hemisphere Temperature Anomalies from 1850-2018 -- Exponential Fit')

def func(x,a,b,c):
    return a*np.exp(-b*x)+c

startyear=1850
endyear=2019
list1=np.arange(0,(endyear-startyear),1)

popt,pcov=curve_fit(func,list1,
                    sh['Median (℃)'].values.tolist(),p0=(1,1e-6,1))
print(popt)
plt.plot(list1,sh['Median (℃)'], '-')
plt.plot(list1,func(list1, *popt),'b--')
```
